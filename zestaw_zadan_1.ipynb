Napisz dataclass o nazwie Measurement, która przechowuje informacje o pojedynczym pomiarze:

    czas pomiaru,

    jednostkę.

Następnie napisz program, który:

    tworzy kilka obiektów tej klasy,
    przechowuje je w liście,
    wypisuje wszystkie pomiary,
    znajduje pomiar o największej wartości.

Można się nauczyć

    dekorator @dataclass

```python
from dataclasses import dataclass, field
```

```python
@dataclass(slots = True)
class Measurement:
    time: float
    value: float
    unit: str 
```

```python
measure = Measurement(time=12.5, value=100.0, unit='Celsius')
print(measure.time)
```

```python
#tworze kilka obiektów klasy Measurement
measure1 = Measurement(time=10.0, value=25.0, unit='Celsius')
measure2 = Measurement(time=15.0, value=30.0, unit='Celsius')
measure3 = Measurement(time=20.0, value=28.0, unit='Celsius')

#przechowuje w liscie te obiekty 
measurements = [measure1, measure2, measure3]

#wypisuje wszystkie pomiary 
for m in measurements:
    print(f'Time: {m.time} s, Value: {m.value} {m.unit}')

#max pomiar wartosci 

maksymalne_czas = max(measurements, key = lambda m: m.time) #lambda bierze z listy i przechodzi po kazdym m 
print(f'Measurement, który ma najwiekszy czas: {maksymalne_czas}')

maksymalne_value = max(measurements, key=lambda m: m.value)
print(f'Measurement, który ma najwieksza wartosc: {maksymalne_value}')
```

Klasa powinna umożliwiać:

    utworzenie wektora o zadanych współrzędnych,
    wypisanie go w czytelnej postaci,
    dodawanie dwóch wektorów operatorem +,

    mnożenie wektora przez liczbę.

Można się nauczyć

    __str__ lub __repr__

    sens metod specjalnych

Rozszerzenie

Dodać metodę obliczającą iloczyn skalarny dwóch wektorów.


```python
import math
class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other:object):
        print("Adding two vectors")
        return Vector2D(self.x + other.x, self.y + other.y)
    
    def dlugosc(self):
        return math.sqrt(self.x**2 + self.y**2)
    
    def __mul__(self, skalar):
        return Vector2D(self.x*skalar, self.y*skalar)
    
    def __rmul__(self, skalar):
        return self.__mul__(skalar)
    
    def __repr__(self):
        return f"Vector2D({self.x}, {self.y})"
    
v1 = Vector2D(5.0, 2.0)
v2 = Vector2D(0.8,4.2)
print(v1 + v2)
print((2*v1))
print(f"Dlugosc v1: {v1.dlugosc()}")
print(f"Dlugosc v2: {v2.dlugosc()}")

v_ska=(v1*3)
print(v_ska)
```

```python
vector = Vector2D(3.0, 4.0)
print(f"Vector: ({vector.x}, {vector.y})")
```

Zadanie 4. Pojemnik na wyniki pomiarów - indeksowanie i iteracja
Cel

Pokazać, że własny obiekt może działać jak kolekcja.
Treść

Napisz klasę MeasurementSeries, która przechowuje serię wyników pomiarowych.

Klasa powinna:

    przyjmować listę liczb przy tworzeniu obiektu,
    pozwalać pobierać elementy przez [],
    pozwalać zmieniać elementy przez [],
    zwracać liczbę elementów przez len(...),
    umożliwiać iterację w pętli for,
    mieć metodę liczącą średnią arytmetyczną.

Można się nauczyć

    __getitem__
    __setitem__
    __len__
    __iter__

Rozszerzenie

Dodać metodę zwracającą największy i najmniejszy pomiar.

```python
class MeasurementSeries:
    def __init__(self, data):
        self.data = list(data) 
    
    #pobieram elementy przez []
    def __getitem__(self, elementy):
        return self.data[elementy]

    def __setitem__(self, elementy, indeks):
        self.data[elementy] = indeks
    
#zwracam liczbe elementów przez len()
    def __len__(self):
        return len(self.data)
    
    #iteracja w petli for 
    def __iter__(self):
        return iter(self.data)
    
    #metoda liczaca srednia z pomiarow 
    def average(self):
        if len(self.data) == 0:
            return 0
        else:
            return sum(self.data) / len(self.data)
    
    #zwracanie najwiekszy i najmniejszy pomiar
    def max(self):
        if len(self.data)==0:
            return None 
        else:
            return max(self.data)
    
    def min(self):
        if len(self.data)==0:
            return None 
        else:
            return min(self.data)
```

```python
measurement_series = MeasurementSeries([10.0, 20.0, 30.0, 40.0])
print(measurement_series[2])
print(f'Length of series: {len(measurement_series)}')
print(measurement_series.average())
print(measurement_series.max())
print(measurement_series.min())

```

Zadanie 5. Funkcja wielomianowa - obiekt funkcyjny
Cel

Implementacja __call__.
Treść

Napisz klasę Polynomial, która reprezentuje wielomian jednej zmiennej. Współczynniki wielomianu powinny być przekazywane przy tworzeniu obiektu w postaci listy.

Przykład: lista [1, 0, 3] oznacza wielomian:

1 + 0 x + 3 x 2

Klasa powinna:

    przechowywać współczynniki,
    pozwalać wypisać wielomian w czytelnej postaci,
    umożliwiać wywołanie obiektu jak funkcji, np. p(2),
    mieć metodę zwracającą stopień wielomianu.

Można się nauczyć

    __call__
    obiekt funkcyjny

Rozszerzenie

Dodać metodę zwracającą wartość pochodnej wielomianu w punkcie.

```python
class Polynomial:
    def __init__(self, wspolczynniki):
        self.wspolczynniki = list(wspolczynniki)
    
    def __repr__(self):
        lista = [f"{coef}*x^{i}" for i, coef in enumerate(self.wspolczynniki)]
        return " + ".join(lista)
    
    #co robi call ?
    # pozwala na wywolanie obiektu klasy jak funkcji, czyli polynom(2) bedzie obliczac wartosc wielomianu dla x=2
    def __call__(self, x):
        wynik =0 
        for i, coef in enumerate(self.wspolczynniki):
            wynik += coef * (x ** i)
        return wynik
    
    #zwraca stopien wielomianu 
    def degree(self):
        return len(self.wspolczynniki) - 1 #bo indeksy zaczynaja sie od 0, wiec stopien to liczba wspolczynnikow -1
    
    #pochodna wielomianu w punkcie 
    def derivative(self, x):
        wynik = 0
        for i, coef in enumerate(self.wspolczynniki):
            if i > 0: #inaczej to stala, pochodna zero 
                wynik += i * coef * (x ** (i - 1)) #wzor na pochodna 
        return wynik
    
    
```

```python
polynom = Polynomial([1,12,-3])
print(polynom.degree())
print(polynom.derivative(1)) #pochodna w punkcie 1
print(polynom(2)) #podstaeia 2 pod x 
print(polynom.wspolczynniki)
```

Pierwsze sensowne dziedziczenie.
Treść

Napisz klasę bazową Account, która reprezentuje konto użytkownika. Powinna przechowywać nazwę użytkownika i liczbę punktów.

Klasa powinna mieć metody:

    dodawania punktów,
    odejmowania punktów,
    wypisywania informacji o koncie.

Następnie napisz klasę PremiumAccount, dziedziczącą po Account, która dodatkowo:

    przechowuje współczynnik premii,
    przy dodawaniu punktów dolicza bonus,
    korzysta z super() tam, gdzie ma to sens.

Można się nauczyć

    dziedziczenie
    nadpisywanie metod
    super()

Rozszerzenie

Dodać trzeci typ konta, np. PenaltyAccount, które przy każdej operacji pobiera niewielką opłatę w punktach.


```python
class Account:
    def __init__(self,nazwa='', punkty=0):
        self.nazwa = nazwa
        self.punkty = punkty 
        print('Creating Account')
    
    def dodanie(self, kwota):
        self.punkty += kwota
    
    def odejmowanie(self, kwota):
        self.punkty -= kwota
    
    def stan_konta(self):
        return self.punkty
```

```python
acc = Account('Jan Kowalski', 100)
print(acc.stan_konta())
```

```python
class PremiumAccount(Account):
    def __init__(self, nazwa='', punkty=0, premia=0.25):
        super().__init__(nazwa, punkty) #klasa nadrzedna 
        self.premia = premia
        #przy dodawaniu punktow, dodaje tez premie
    def dodanie(self, kwota):
        bonus = kwota * self.premia 
        super().dodanie(kwota + bonus) #dodanie w klasie nadrzednej  
        
        
```

```python
PA = PremiumAccount('Anna Nowak', 200,0.25) #nie zapomniec o premii 
print(PA.stan_konta())
print('Dodaje premie')
PA.dodanie(100) 
print(f'Anna Nowak ma teraz premie: {PA.stan_konta()}') #200 + 100 + 25% z 100 = 325
```

```python
PA = PremiumAccount('Anna Nowak', 200, 0.25)

print(PA.stan_konta())   # 200
PA.dodanie(100)
print(PA.stan_konta()) 
```

Zadanie 7. Zwierzęta na planszy - polimorfizm
Cel

Pokazać wspólny interfejs i różne implementacje.
Treść

Napisz klasę bazową Animal, która przechowuje imię oraz pozycję zwierzęcia na planszy (x, y).

Klasa powinna mieć metody:

    move()
    sound()
    info()

Następnie napisz co najmniej dwie klasy dziedziczące, np.:

    Dog
    Bird

Każda z nich powinna:

    inaczej implementować metodę sound(),
    inaczej implementować metodę move(),
    korzystać z konstruktora klasy bazowej przez super().

Na końcu napisz program, który tworzy listę różnych zwierząt i wywołuje na nich te same metody.
Można się nauczyć

    polimorfizm
    wspólny interfejs
    praktyczny sens klasy bazowej

Rozszerzenie

Niech klasa Animal będzie zaimplementowana jako abstrakcyjna.

```python
class Animal:
    def __init__(self, name='', x=0, y=0):
        self.name = name
        self.x = x
        self.y = y
    def move(self, ruch_w_osi_x, ruch_w_osi_y):
        self.x += ruch_w_osi_x
        self.y += ruch_w_osi_y
        print(f'{self.name} moved to ({self.x}, {self.y})')
    
    def sound(self):
        print(f'{self.name} robi dzwiek')
    
    def info(self):
        print(f'Name: {self.name}, Position: ({self.x}, {self.y})')
```

```python
ani = Animal('Kon Lisa', 3, 6)
ani.move(5, 3)

ani.info()
```

```python
class Dog(Animal):
    def sound(self):
        print(f'{self.name} says: Woof!')
    
    def move(self, ruch_w_osi_x, ruch_w_osi_y):
        super().move(ruch_w_osi_x, ruch_w_osi_y) #wywoluje metode move z klasy nadrzednej 
        print(f'{self.name} rusza ogonem, gdy rusza sie !')

class Cat(Animal):
    def sound(self):
        print(f'{self.name} says: Meow!')
    
    def move(self, ruch_w_osi_x, ruch_w_osi_y):
        super().move(ruch_w_osi_x, ruch_w_osi_y) #wywoluje metode move z klasy nadrzednej 
        print(f'{self.name} miauczy ruszajac sie !')
```

```python
doggo = Dog('Rex', 1, 3)
doggo.sound()
doggo.move(2, 2)
print(f'Rex moves, new position: {doggo.x}, {doggo.y}')
doggo.info()
```

```python
catto = Cat('Mia', 0, 4)
catto.sound()
catto.move(1, 1)
print(f'Mia moves, new position: {catto.x}, {catto.y}')
catto.info()
```

```python
#lista roznych zwierzat i wykonuje metode 
animals = [Dog('Rex', 1, 3), Cat('Mia', 0, 4), Animal('Gepard', 2, 2)]
for animal in animals:
    animal.sound()
    
```

```python
for animal in animals:
    animal.move(1, 1)
```
